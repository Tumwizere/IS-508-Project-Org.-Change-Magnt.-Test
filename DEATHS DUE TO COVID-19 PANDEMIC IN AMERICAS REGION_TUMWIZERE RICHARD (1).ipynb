# Simple Linear Regression

y=mx+c 

y= independent_variable , m= slope , x= independent_variable , c=y-intercept

```python
from sklearn.linear_model import LinearRegression
```

```python
#Using identified cases to predict deaths in the next 24 hours
x = richard[['Cases - newly reported in last 24 hours']] 
y = richard['Deaths - newly reported in last 24 hours'] 
lm = LinearRegression()

lm.fit(x,y)

Yhat=lm.predict(x)
Yhat[0:5]
```

```python
# Calculationg Coeficient
lm.coef_
```

```python
#Intercept
lm.intercept_
```

```python
#calculatin R-Squared
lm.score(x,y)
```

```python
#Calculating R-squared
print('The R-squared is: ', lm.score(x, y))
```

```python
from sklearn.metrics import mean_squared_error #for calculating MSE
from math import sqrt 
```

```python
#calculating MSE
y=lm.predict(x)
MSE = mean_squared_error (richard['Deaths - newly reported in last 24 hours'], y)
print('MSE is', MSE)
```

```python
#Calculating Root mean squared error
print('Root Mean Squared Error:',sqrt(MSE))
```
